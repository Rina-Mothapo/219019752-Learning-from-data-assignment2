<h2> Processing the data </h2>

```python
##importing the library that will help with the cleaning and visualization of the data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
##importing the dataset needed in this analysis
dataset1 = pd.read_excel('219019752 Learning from data (1).xlsx')
```

```python
dataset1
```

<h2> Cleaning the data </h2>

```python
##deleting the timestamp column as it is not needed for the analysis
dataset1 = dataset1.drop(['Timestamp'], axis = 1)
```

```python
dataset1
```

```python
##checking if there are any NaN values and if there are, how many in each column?
dataset1.isna().sum()
```

```python
##deleting the rows with NaN values
dataset1 = dataset1.dropna()
```

```python
##checking to confirm that the rows with NaN values are deleted
dataset1.isna().sum()
```

```python
dataset1
```

<h4>Checking for duplicates</h4>


```python
dataset1.duplicated().sum()
```

<h4>Deleting duplicates</h4>

```python
dataset1 = dataset1.drop_duplicates()
```

```python
dataset1.duplicated().sum()
```

<h4>Renaming columns</h4>

```python
dataset1 = dataset1.replace('self_employed', 'Self_employed')
dataset1 = dataset1.replace('family_history', 'Family_history')
dataset1 = dataset1.replace('treatment', 'Treatment')
dataset1 = dataset1.replace('mental_health_interview', 'Mental_health_interview')
```

```python
dataset1 = dataset1.replace('No', 1)
dataset1 = dataset1.replace('Maybe', 2)
dataset1 = dataset1.replace('Yes', 3)
dataset1 = dataset1.replace('Not sure', 4)

dataset1 = dataset1.replace('1-14 days', 14)
dataset1 = dataset1.replace('Go out Every day', 13)
dataset1 = dataset1.replace('More than 2 months', 17)
dataset1 = dataset1.replace('15-30 days',15 )
dataset1 = dataset1.replace('31-60 days', 16)
```

<h3>Descriptive statisctics</h3>

<h4>Central tendency measurements</h4>

```python
mean= dataset1['family_history'].mean()
median = dataset1['family_history'].median()
mode = dataset1['family_history'].mode()

print("The mean is " + str(mean) + ", the median is " + str(median) + " and the mode is " + str(mode))
```

<h4>Observations on central measurements</h4>
On average 1.83 of the candidates have a family history of mental health issues, the middle value of candidates is 1.0, and the the most frequent value in the dataset is 1. (Please refer to the legend at the top of the notebook to further understand the value interpretations)




<h4>Dispersion measures</h4>


```python
variance = dataset1['care_options'].var()
std_deviation = dataset1['care_options'].std()
iqr = dataset1['care_options'].quantile(0.75) - dataset1['care_options'].quantile(0.25)

print("The variance value is " + str(variance) + ", the standard deviation is " + str(std_deviation) + ", and the interquatile range is " + str(iqr))
```

<h4>Distribution and shape of the dataset</h4>

```python
skewness = dataset1['Mental_Health_History'].skew()
kurtosis = dataset1['Mental_Health_History'].kurtosis()

print("The skewness of the dataset is " + str(skewness) + ", and the kurtosis " + str(kurtosis))
```
